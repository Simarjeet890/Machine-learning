```python
#this model uses the decision tree algorithm to classify data
from sklearn.tree import DecisionTreeClassifier
#use train_test_split to split the data into training and testing sets
from sklearn.model_selection import train_test_split
#accuracy score to evaluate the model. the accuracy score tests TP and TN against total predictions
from sklearn.metrics import accuracy_score
model = DecisionTreeClassifier()
import pandas as pd 
#joblib works with large datasets, used to dump file
#pip install
import joblib
 
```

```python
#pandas is used to import comma separated values file data. turns it into dataframe
music_data = pd.read_csv('music.csv')
music_data
```

```python
#genre is dropped from the datasets to get input variable gender and age
X = music_data.drop(columns=['genre'])
#genre is the output variable we are trying to predict
y = music_data['genre']
```

```python
#sklearn.model_selection module uses train_test_split to split data 80/20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#fit() method trains the model on the x and y training sets
model.fit(X_train, y_train)
#dumps the trained model into a joblib file
joblib.dump(model, 'music_model.joblib')
```

```python
#makes the predictions based off age-gender pair
model.predict([[21,1]])
```

```python
predictions = model.predict(X_test)
#accuracy score is calculated using the y_test 
score = accuracy_score(y_test, predictions)
score
```
